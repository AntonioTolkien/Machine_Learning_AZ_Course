# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
# Obtener la pendiente (coeficiente) del modelo
slope = regressor.coef_

# Obtener el término de intersección (intercepto) del modelo
intercept = regressor.intercept_

print(f"Pendiente (m): {slope[0]}")
print(f"Término de intersección (b): {intercept}")
print(f"La ecuación de la regresión lineal es: y = {slope[0]}X + {intercept}")

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.plot(X_train, regressor.predict(X_train), color='red', label='Línea de regresión')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos de prueba')
plt.plot(X_test, regressor.predict(X_test), color='red', label='Línea de regresión')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```
